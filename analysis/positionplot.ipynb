```python
df=pd.read_csv("wgene183.csv")
order_or=["Passeriformes","Psittaciformes","Falconiformes","Coraciiformes","Piciformes","Bucerotiformes","Coliiformes","Strigiformes","Accipitriformes","Pelecaniformes","Procellariiformes","Sphenisciformes","Phaethontiformes","Eurypygiformes","Charadriiformes","Gruiformes","Caprimulgiformes","Otidiformes","Musophagiformes","Cuculiformes","Mesitornithiformes","Columbiformes",'Podicipiformes',"Galliformes","Anseriformes"]
ref=pd.read_csv("sle_104.csv",header=0,index_col=0)
for k,row in df.iterrows():
    df.loc[k,'order']=ref.loc[df.loc[k,'b10kid'],'order']
    df.loc[k,'subfamily']=ref.loc[df.loc[k,'b10kid'],'subfamily']
```

```python
df.fillna(float(0),inplace=True)
df['order'] = df['order'].astype('category')
df['order'].cat.reorder_categories(order_or, inplace=True)
df.sort_values('order', inplace=True)
df
```

```python
def xs(ar):
    list1=[]
    flag=0
    t=[]
    count1=0
    count2=0
    for k in ar:
        if (k==5) & (flag==0):
            list1.append(t)
            count1=count1+len(t)
            t=[k]
            flag=1
            continue
        elif (k==5) & (flag==1):
            t.append(k)
            continue
        elif (k!=5) & (flag==1):
            list1.append(t)
            count2=count2+len(t)
            t=[k]
            flag=0
            continue
        elif (k!=5) & (flag==0):
            t.append(k)
            continue
    ok=[]
    rem=1000-count1
    fold=count2/rem
    for i in list1:
        if 5 in i:
            l2=int(len(i)/fold)
            op=list(np.ones(l2)*5)
        else:op=i
        ok=ok+op
    return ok
```

```python
df=pd.read_csv("wgene2183.csv")
cluster1=["Passeriformes","Psittaciformes","Falconiformes","Coraciiformes","Piciformes","Bucerotiformes","Coliiformes","Strigiformes","Accipitriformes"]
cluster2=["Pelecaniformes","Procellariiformes","Sphenisciformes","Phaethontiformes","Eurypygiformes"]
cluster3=["Charadriiformes","Gruiformes","Opisthocomiformes","Caprimulgiformes","Otidiformes","Musophagiformes","Cuculiformes"]
cluster4=["Mesitornithiformes","Columbiformes","Podicipediformes"]
gns=list(df.columns[51:235])
```

```python
dicto={}
cldf=[cluster1,cluster2,cluster3,cluster4]
for i in range(0,4):
    cn=[]
    cdf=df[df.apply(lambda x: x['order'] in cldf[i], axis=1)]
    for j in gns:
        n=len(np.unique(cdf.loc[cdf[j]==1]['order']))
        cn.append(n)
    dicto.update({"cluster"+str(i+1):cn})
countdf=pd.DataFrame(dicto,index=gns)
countdf
```
