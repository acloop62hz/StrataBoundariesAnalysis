```python
import pandas as pd
import csv
```

```python
df=pd.read_csv("w.gene.csv",header = 0)
df.drop(index=0,inplace=True)
spes=['NCBI-021', 'NCBI-018', 'APP-023', 'NCBI-012','PAS-001', 'PAS-020', 'B10K-DU-002-22', 'PAS-017', 'PAS-018', 'PAS-019','APP-016', 'GAL-008', 'ANS-002']
df['0']=df['0'].str.upper()
df
```

```python

edict={'bid':[],'gname':[],'gid':[]}
c=pd.DataFrame(edict)
for i in spes:
    dft=df.loc[df[i]!='-']
    for k,row in dft.iterrows():
        gname=dft.loc[k,'0']
        gid=dft.loc[k,i]
        c.loc[len(c)]={'bid':i,'gname':gname,'gid':gid}

```

```python
c.dropna(axis=0,how='any',inplace=True)
c.to_csv('add_verified2.csv',index=False)
```
